### 1) Scatterplot (Numerical-Numerical)

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'])
```

```python
# finding out the gender of the customers
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'])
```

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'], style=tips['smoker'])
```

### 2) Bar plot (Numerical - Categorical)

```python
titanic.head()
```

```python
# we generally keep the categorical data on the x axis and the numerical one on the y-axis
sns.barplot(x=titanic['Pclass'], y=titanic['Age'])
```

```python
sns.barplot(x=titanic['Pclass'], y=titanic['Fare'], hue=titanic['Sex'])
```

### 3) Box plot (Numerical-Categorical)

```python
sns.boxplot(x='Sex', y='Age', hue='Survived', data=titanic)
```

### 4) Distplot (Numerical-Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
```

### 5) Heatmap (Categorical-Categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

Color indicates strength and direction:

🔴 Bright red: Strong positive correlation (e.g., Fare and Age)

🔵 Bright blue: Strong negative correlation (e.g., Pclass and Fare)

⚪ White or pale colors: Near-zero correlation (e.g., Age and Survived)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']),annot=True)
```

```python
(titanic.groupby('Pclass')['Survived'].mean() * 100).plot(kind='bar')
```

```python
(titanic.groupby('Embarked')['Survived'].mean() * 100).plot(kind='bar')
```

### 6) Clustermap (Categorical-Categorical)

```python
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

### 8) Lineplot (Numerical-Numerical)

A lineplot is one where we join all the points of a scatterplot using a line. We should use it mainly when the quantity on our x axis is of time category

```python
flights.head()
```

```python
new=flights.groupby('year')['passengers'].sum().reset_index()
```

```python
sns.lineplot(x=new['year'], y=new['passengers'])
```

```python
flights.pivot_table(values='passengers', index='month', columns='year')
```

```python
sns.heatmap(flights.pivot_table(values='passengers', index='month', columns='year'))
```
