## Voting Ensemble 



Take mode in classification, and mean in regression

The models should be quite dissimilar for more accuracy. Also, no individual model should have an accuracy less than 50%

## Why does voting work to give a model more accurate than individual model?

It is due to probability. As we are making the models independent and different, probabilities get multiplied and added making the probability higher

If the individual models have less than 50% accuracy, then it will get snowballed and make our final ensemble have accuracy lower than the worst individual model
