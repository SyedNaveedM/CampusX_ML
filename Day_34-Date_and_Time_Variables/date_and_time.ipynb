### Working with dates

```python
# converting to datetime datatype
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

### 1) Extract year

```python
date['year']=date['date'].dt.year

date.sample(5)
```

### 2) Extract month

```python
# month() will give the month number, like 1 for january
date['month']=date['date'].dt.month_name()

date.sample(5)
```

```python
date['month_no']=date['date'].dt.month

date.sample(5)
```

### 3) Extract days

```python
date['day']=date['date'].dt.day
```

```python
date.sample(5)
```

```python
# day name
date['day_name']=date['date'].dt.day_name()

date.sample(5)
```

```python
# day of week
date['day_week']=date['date'].dt.day_of_week
```

```python
date.sample(5)
```

### 4) Is weekend?

```python
# is weekend?

date['date_is_weekend'] = np.where(date['day_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
date['quarter']=date['date'].dt.quarter

date.sample(5)
```

### Extract Semester

```python
date['semester']=np.where(date['quarter'].isin([1,2]), 1, 2)

date.sample(5)
```

### Extract time elapsed between dates

```python
import datetime as dte

today=dte.datetime.today()
```

```python
today
```

```python
today-date['date']
```

```python
# no of days elapsed
(today-date['date']).dt.days
```

```python
# months passed
np.round((today - date['date']) / np.timedelta64(30, 'D'), 0)
```
