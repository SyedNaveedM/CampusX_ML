### üß† Kernel Trick in Support Vector Machines (SVM)

The **kernel trick** is a powerful technique that allows SVMs to solve **non-linearly separable problems** by **implicitly mapping data into higher-dimensional space** ‚Äî *without actually computing the transformation*.

---

## üîπ Motivation

### üü¢ Linear SVM:

Works only if data is linearly separable.

### üî¥ Real-world data:

Often **not linearly separable** in input space.

### ‚úÖ Solution: Map to higher dimensions where separation is possible.

---

## üî∏ Key Idea

Let $\phi(x)$ be a mapping function that transforms input data into higher-dimensional space.

Instead of explicitly computing:

```math
\phi(x)^T \phi(x')
```

...we use a **kernel function**:

```math
K(x, x') = \phi(x)^T \phi(x')
```

This saves time and memory, especially in **very high or infinite-dimensional spaces**.

---

## üîπ What Is a Kernel?

A **kernel** is a function that computes the **dot product in some feature space** ‚Äî **without** knowing the feature space.

```math
K(x, x') = \langle \phi(x), \phi(x') \rangle
```

---

## üî∏ Common Kernel Functions

| Kernel Name        | Formula                                 | Description                                    |
| ------------------ | --------------------------------------- | ---------------------------------------------- |
| **Linear**         | $K(x, x') = x^T x'$                     | No transformation ‚Äî regular dot product        |
| **Polynomial**     | $K(x, x') = (x^T x' + c)^d$             | Allows curved boundaries, degree = flexibility |
| **RBF / Gaussian** | $K(x, x') = \exp(-\gamma \|x - x'\|^2)$ | Infinite-dimensional, radial separation        |
| **Sigmoid**        | $K(x, x') = \tanh(\alpha x^T x' + c)$   | Like neural networks                           |

---

## üîπ Example: Non-linear ‚Üí Linear with Kernel

Suppose data is not linearly separable in 2D:

```python
from sklearn.svm import SVC

model = SVC(kernel='rbf', C=1.0, gamma='scale')
model.fit(X_train, y_train)
```

* SVM **thinks** it's operating in higher dimensions
* But thanks to the **kernel trick**, we never compute $\phi(x)$

---

## üî∏ Benefits of Kernel Trick

‚úÖ Avoids explicit feature transformation
‚úÖ Handles complex, non-linear boundaries
‚úÖ Enables efficient computation in **very high-dimensional** feature spaces
‚úÖ Flexible modeling using just a kernel function

---

## üî∏ Drawbacks

‚ùå Requires careful tuning (`C`, `gamma`, kernel type)
‚ùå Can be **computationally expensive** on large datasets
‚ùå Choice of kernel highly impacts performance

---

## üß† Summary

| Concept      | Description                                         |
| ------------ | --------------------------------------------------- |
| Kernel Trick | Avoids direct computation of feature transformation |
| Used In      | SVM, Kernel Ridge, Gaussian Processes, PCA          |
| Best Kernel? | Depends on the **data distribution** and **task**   |

---



````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.axes._axes import _log as matplotlib_axes_logger
from mpl_toolkits import mplot3d
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from matplotlib.colors import ListedColormap
````

````python
from sklearn.datasets import make_circles
X, y = make_circles(100, factor=.1, noise=.1)

plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='bwr')
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
````

````python
classifier = SVC(kernel="linear")
classifier.fit(X_train, y_train.ravel())
y_pred = classifier.predict(X_test)
````

````python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
````

````python
zero_one_colourmap = ListedColormap(('blue', 'red'))
def plot_decision_boundary(X, y, clf):
    X_set, y_set = X, y
    X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, 
                                 stop = X_set[:, 0].max() + 1, 
                                 step = 0.01),
                       np.arange(start = X_set[:, 1].min() - 1, 
                                 stop = X_set[:, 1].max() + 1, 
                                 step = 0.01))
  
    plt.contourf(X1, X2, clf.predict(np.array([X1.ravel(), 
                                             X2.ravel()]).T).reshape(X1.shape),
               alpha = 0.75, 
               cmap = zero_one_colourmap)
    plt.xlim(X1.min(), X1.max())
    plt.ylim(X2.min(), X2.max())
    for i, j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = (zero_one_colourmap)(i), label = j)
    plt.title('SVM Decision Boundary')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.legend()
    return plt.show()
````

````python
plot_decision_boundary(X, y, classifier)
````

````python
def plot_3d_plot(X, y):
    r = np.exp(-(X ** 2).sum(1))
    ax = plt.subplot(projection='3d')
    ax.scatter3D(X[:, 0], X[:, 1], r, c=y, s=100, cmap='bwr')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')
    ax.set_zlabel('y')
    return ax
````

````python
plot_3d_plot(X,y)
````

````python
rbf_classifier = SVC(kernel="rbf")
rbf_classifier.fit(X_train, y_train)
y_pred = rbf_classifier.predict(X_test)
````

````python
accuracy_score(y_test, y_pred)
````

````python
plot_decision_boundary(X, y, rbf_classifier)
````

````python
poly_classifier = SVC(kernel="poly",degree=2)
poly_classifier.fit(X_train, y_train)
y_pred = poly_classifier.predict(X_test)
````

````python
accuracy_score(y_test, y_pred)
````

````python
plot_decision_boundary(X, y, poly_classifier)
````
