A point is considered an outlier if it lies far from the central tendency (mean, median) or outside the expected distribution.

Outliers will distract the model by a lot sometimes so they have to be taken care of carefully

### When is outlier dangerous?

For example in anomaly detection algorithms, our goal is to catch the outliers so we need them. Outliers are dangerous dependent on the problem statement, algorithm, etc.

Linear regression, Logistic regression and adaboost and deep learning algortihms are very badly affected by outliers as they all calculate weights.

Non-weight algorithms like decision trees are unaffected by outliers

### How to treat outliers?

1) Trimming: Just remove the outlier data points. This can be harmful when there are many outliers
2) Capping: Keeping a cap on how much range of the data we consider normal
3) Treat them as missing values

### How to detect outliers?

All the observations out of range $\mu \pm 3\sigma$ can be treated as outliers

Use boxplot to get the outliers

Use percentiles for detecting them

### Techniques for removal of outliers:

1) Z-score treatment
2) IQR based filtering
3) Percentile
4) Winsorization
