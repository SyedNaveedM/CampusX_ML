### Encoding numerical features

Sometimes encoding the numerical values is better representation than purely numerical

Ex: Downloads of an app. It is better to make categories/binning of the no of downloads in a range helps make our problem easier

Two ways to do it:
1) Discretization/Binning
2) Binarization

### Discretization

Types of discretization:
1) Unsupervised binning
2) Supervised binning
3) Custom binning

Unsupervised binning consists of:
1) Equal width (uniform)
2) Equal frequency (quantile)
3) K-Means binning

Supervised binning has only decision tree binning

### Equal width/uniform binning

We first have to decide the no of bins

Then we do (max-min)/bin to get the no of intervals and do (start, start+offset) and so on

The spread of the data is not changed and also outliers are handled better

### Equal frequency/quantile binning

We use quantiles here to get the equal frequency

It makes the spread of the data uniform and handles outliers well. This is why this is used more

### K-Means binning

This is used mainly when we have our data in clusters

The interval is called centroids

We map the datapoints which are closest to the centroids in that cluster
