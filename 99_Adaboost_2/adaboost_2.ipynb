Here, error is the same as linear regression

Error is the sum of the weights of the misclassified samples

We increase the importance of the samples that were misclassified. So, we use upsampling that is increasing the weights of the samples



The weights have to be always normalized



We then calculate ranges of the weights. Next, we generate 5 random nos from 0 and 1 and pick the row in that range. As we increased the weight of the rows that were misclassified, the probability of picking those rows increases. This is called upsampling
