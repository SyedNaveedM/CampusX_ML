## Logistic Regression

The main requirement for applying this is that the data is linearly separable

### Perceptron Trick

This trick doesn't give the best possible solution for logistic regression problem

In logistic regression we take the equation of the line as ax+by+c=0

To determine the labels, we simply subsitute the particular point in the line's equation. If it is > 0, then it is +ve region otherwise -ve

If we want to shift a point to the negative region, then we do (a-x1,b-y1,c-1) and create the new line with these coefficients. Add if we want to shift it to the positive region

We add a column full of 1's and keep the model as sigma(wi*xi)=0

### Algorithm

1) Decide no of epochs
2) Select a learning rate eta
3) We select a point at random
4) If the point belongs to negative class and the model shows positive then we do w_new=w_old - eta*(point). We add if it is the opposite
