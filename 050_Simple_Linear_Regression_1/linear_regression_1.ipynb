Linear regression is a supervised regression algorithm

Types of linear regression:
1) Simple - 1 feature input, 1 target variable
2) Multiple - multiple features as input, 1 target variable
3) Polynomial

### Simple Linear Regression

We have cgpa as the input and the output as the package

When the points are plotted using a scatterplot (the linear points are just demonstration of a perfectly linear dataset which is unlikely in real world datasets)

The best fit line - the line which has the least deviation or distance from all the points in the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('placement.csv')
```

```python
df.sample(3)
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
X=df[['cgpa']]
y=df[['package']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred=lr.predict(X_test)

print("MSE:", mean_squared_error(y_test, y_pred))
print("R² Score:", r2_score(y_test, y_pred))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```
