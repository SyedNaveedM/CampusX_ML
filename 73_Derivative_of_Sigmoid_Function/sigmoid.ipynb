## ðŸ§  Definition of Sigmoid Function

The sigmoid (logistic) function is defined as:

```math
Ïƒ(x) = \frac{1}{1 + e^{-x}}
```

It maps any real number to the range (0, 1).

---

## ðŸŽ¯ Goal

Find the derivative:

```math
\frac{dÏƒ(x)}{dx}
```

---

## ðŸ”„ Step-by-Step Derivation

### Step 1: Rewrite Ïƒ(x) for differentiation

Let:

```math
Ïƒ(x) = y = \frac{1}{1 + e^{-x}}
```

Differentiate y with respect to x:

```math
\frac{dy}{dx} = \frac{d}{dx} \left[\frac{1}{1 + e^{-x}}\right]
```

---

### Step 2: Use the chain rule and quotient rule

Let:

* `u(x) = 1`
* `v(x) = 1 + e^{-x}`

Then using the quotient rule:

```math
\frac{dy}{dx} = \frac{v \cdot \frac{du}{dx} - u \cdot \frac{dv}{dx}}{v^2}
```

Since:

* `du/dx = 0`
* `dv/dx = -e^{-x}`

We get:

```math
\frac{dy}{dx} = \frac{0 - (1)(-e^{-x})}{(1 + e^{-x})^2} = \frac{e^{-x}}{(1 + e^{-x})^2}
```

---

### Step 3: Express in terms of Ïƒ(x)

Recall:

```math
Ïƒ(x) = \frac{1}{1 + e^{-x}} \Rightarrow 1 - Ïƒ(x) = \frac{e^{-x}}{1 + e^{-x}}
```

Now multiply:

```math
Ïƒ(x) (1 - Ïƒ(x)) = \left(\frac{1}{1 + e^{-x}}\right) \left(\frac{e^{-x}}{1 + e^{-x}}\right) = \frac{e^{-x}}{(1 + e^{-x})^2}
```

Which matches our earlier result.

---

## âœ… Final Result

```math
\frac{dÏƒ(x)}{dx} = Ïƒ(x)(1 - Ïƒ(x))
```
