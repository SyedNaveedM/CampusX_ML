### 🧠 KNN Imputation: How to Impute Missing Values from Nearest Neighbors

Once you have identified the **k nearest neighbors** for a sample with missing values (e.g., missing `Age`), the imputation method depends on the **type of feature**:

---

### 🔢 Numerical Features (e.g., Age, Income)

You typically take the **average** of the known values from the `k` neighbors.

**🎯 Formula:**

$$
\text{Imputed value of } f = \frac{1}{k} \sum_{i=1}^{k} f_i
$$

Where:

* $f_i$ is the value of feature $f$ in the $i^{\text{th}}$ neighbor

📌 **Example:**
If two closest neighbors (row 3 and 4) have `Age = [35, 45]`, then:

$$
\text{Imputed Age} = \frac{35 + 45}{2} = 40
$$

---

### 🔠 Categorical Features (e.g., Gender, Occupation)

You typically take the **mode** (most frequent category) among the `k` neighbors.

**🎯 Example:**
If the `k` neighbors have:

```
Gender = [Male, Female, Male]
```

Then:

```
Imputed Gender = Male
```

---

### ✅ Summary Table

| Feature Type | Imputation Rule               |
| ------------ | ----------------------------- |
| Numerical    | Mean (or median) of neighbors |
| Categorical  | Mode of neighbors             |


📌 Key Rules:
Always filter out NaN values before calculating mean or mode.

If no neighbors have a valid value for that feature, then:

Try increasing k

Or fallback to global strategies (mean, median, mode)

Advantage: More accurate
Disadvantage: 
1) More no of calculations
2) Takes too much memory as the training set has to be on the server
